# **Canny edge detection is a multi-step algorithm that can detect edges with noise supressed at the same time. Smooth the image with a Gaussian filter to reduce noise and unwanted details and textures.**

```
#import image processing libraries
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```
### Load original image
image = cv2.imread("test_image.jpg")
cv2_imshow(image)
```

```
### define function for canny-edge filter
def canny(img):
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    kernel = 5
    blur = cv2.GaussianBlur(gray,(kernel, kernel),0)
    canny = cv2.Canny(gray, 50, 150)
    return canny
```

```
### return gaussian blur to reduce noise in image
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
kernel = 5
blur = cv2.GaussianBlur(gray,(kernel, kernel),0)
cv2_imshow(blur)
```

```
### return edge filtered image output
image = cv2.imread("test_image.jpg")
lane_image = np.copy(image)
lane_canny = canny(lane_image)
```
